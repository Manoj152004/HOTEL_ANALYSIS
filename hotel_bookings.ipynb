```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\System -02\Downloads\Excel clean_02_hotel_bookings.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
df.drop_duplicates
df.head(2)
```

```python
df.isnull().sum().sort_values(ascending=False).head(5)
```

```python
df[['meal_plan','children','country_of_guest','special_requests']].head(2)
```

```python
for col in['meal_plan','country_of_guest']:
    df[col] = df[col].fillna(df[col].mode()[0])
    
for co in ['children','special_requests']:
    df[co]= df[co].fillna(df[co].median())
```

```python
df.isnull().sum()
```

```python
df.iloc[50:55]
```

```python
df['arrival_date'] = pd.to_datetime(df['arrival_date'],dayfirst=True,errors='coerce')
```

```python
df['children'] = df['children'].astype(int)
```

```python
for i in ['hotel_name','room_type','booking_channel','country_of_guest','cancellation_status']:
    df[i] = df[i].str.title()
```
